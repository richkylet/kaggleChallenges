# San Francisco Crime data

Download data here https://www.kaggle.com/c/sf-crime/data

### Import packages etc

```python
%matplotlib inline
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
sns.set(font_scale=1.5)  
```

### Load data

```python
crime_df = pd.read_csv("/Users/KRich/GitHub/data/SF crime/train.csv")
# remove address, resolution and description
crime_df = crime_df.drop(['Address','Resolution','Descript'],axis=1)
crime_df.head()
```

### Sort data and set the order for plotting

```python
# convert dates to datetime (pd.to_datatime(df))
crime_df['Dates']=pd.to_datetime(crime_df["Dates"])
# year data ---------------------------------
crime_df['Year']  = crime_df['Dates'].dt.year
crime_df['Year']  = crime_df['Year'].astype(str)
# month data (replace with names) ------------d
crime_df['Month'] = crime_df['Dates'].dt.month
month_map = {1:'Jan', 2:'Feb', 3:'Mar', 4:'Apr', 5:'May', 
             6:'Jun', 7:'Jul', 8:'Aug', 9:'Sep', 10:'Oct', 
             11:'Nov', 12:'Dec'}
month_order       = [k for k in month_map.values()] 
crime_df['Month'].replace(month_map, inplace=True)
# Hour data -------------------------------------
crime_df['Hour']  = crime_df['Dates'].dt.hour
crime_df['Hour']  = crime_df['Hour'].astype(str)
# week data
day_order = ['Mon', 'Tue', 'Wed', 
             'Thu', 'Fri', 'Sat', 'Sun']
```

## PLOT

### Average crimes per month every year
Plot average value (barplot) and check for outliers (boxplot)

```python
plt.figure(figsize=(14,10))

# average crimes per month each year
axis1=  plt.subplot2grid((2,2),(0,0))
Month_count = crime_df[["Month", "Category","Year"]].groupby(['Month','Year'],as_index=False).count()
#xxx = xx.groupby(['Year'],as_index=False).mean()
monthbar = sns.barplot(x='Month', y='Category', data=Month_count,palette="Blues_r",
                   order=month_order, ax=axis1)
monthbar.set_xlabel('Month')
monthbar.set_ylabel('Count of crimes')
_ = plt.setp(monthbar.get_xticklabels(), rotation=90)

# boxplot, check for outliers
axis2=  plt.subplot2grid((2,2),(0,1))
monthbox = sns.boxplot(x='Month', y='Category', data=Month_count,palette="Blues_r",
                   order=month_order, ax=axis2)
monthbox.set_xlabel('Month')
monthbox.set_ylabel('Count of crimes')
_ = plt.setp(monthbox.get_xticklabels(), rotation=90)
plt.tight_layout()
```

_Note_: There are outliers in May-July. But, the largest is in May, so we should further investigate May to see if we should exclude data for a given year (assuming data is missing). 

```python
plt.figure(figsize=(14,10))
#=================================================
axis1=  plt.subplot2grid((2,3),(0,0))
mo = 'Apr'
monval = crime_df.loc[crime_df['Month'] == mo]
Month_count = monval[["Category","Year"]].groupby(['Year'],as_index=False).count()
mayplot = sns.barplot(x='Year', y='Category', 
                      data=Month_count,palette="Blues_r", ax=axis1)
mayplot.set_xlabel('Year #')
mayplot.set_ylabel('Count of crimes')
mayplot.set_title(mo)
_ = plt.setp(mayplot.get_xticklabels(), rotation=90)
#=================================================
axis2=  plt.subplot2grid((2,3),(0,1))
mo = 'May'
monval = crime_df.loc[crime_df['Month'] == mo]
Month_count = monval[["Category","Year"]].groupby(['Year'],as_index=False).count()
mayplot = sns.barplot(x='Year', y='Category', 
                      data=Month_count,palette="Blues_r", ax=axis2)
mayplot.set_xlabel('Year #')
mayplot.set_ylabel('Count of crimes')
mayplot.set_title(mo)
_ = plt.setp(mayplot.get_xticklabels(), rotation=90)
#=================================================
axis3=  plt.subplot2grid((2,3),(0,2))
mo = 'Jun'
monval = crime_df.loc[crime_df['Month'] == mo]
Month_count = monval[["Category","Year"]].groupby(['Year'],as_index=False).count()
mayplot = sns.barplot(x='Year', y='Category', 
                      data=Month_count,palette="Blues_r", ax=axis3)
mayplot.set_xlabel('Month')
mayplot.set_ylabel('Count of crimes')
mayplot.set_title(mo)
_ = plt.setp(mayplot.get_xticklabels(), rotation=90)
#=================================================
axis4=  plt.subplot2grid((2,3),(1,0),colspan=3)
# count for 2015 only ---------------
yearval = crime_df.loc[crime_df['Year'] == '2015']
Month_count = yearval[["Month", "Category"]].groupby(['Month'],as_index=False).count()
monthbar = sns.barplot(x='Month', y='Category', data=Month_count,color="royalblue",
                   order=month_order, ax=axis4, label="2015 only")
monthbar.set_xlabel('Month')
monthbar.set_ylabel('Count of crimes')
# average per month-------------
yearval = crime_df.loc[crime_df['Year'] != '2015']
Month_count = crime_df[["Month", "Category","Year"]].groupby(['Month','Year'],as_index=False).count()
monthbar = sns.barplot(x='Month', y='Category', data=Month_count,color="lightgrey",
                   order=month_order, ax=axis4, alpha=.6, label='Avg (2003-2014)')
monthbar.set_xlabel('Month')
monthbar.set_ylabel('Count of crimes')
_ = plt.setp(monthbar.get_xticklabels(), rotation=90)
plt.legend();
#=================================================
plt.tight_layout()
```

May 2015 is suspicious. We can either replace values for this month with average values, drop the month or the year. Considering there is no data after June 2015, lets just drop the year to simplify. 

```python
# remove/drop data from 2015 since its not complete:
crime_df = crime_df[crime_df.Year != '2015']
```

```python
# PLOT
plt.figure(figsize=(14,10))
#=================================================
# average crimes per month each year
axis1=  plt.subplot2grid((2,2),(0,0))
Month_count = crime_df[["Month", "Category","Year"]].groupby(['Month','Year'],as_index=False).count()
monthbar = sns.barplot(x='Month', y='Category', data=Month_count,palette="Blues_r",
                   order=month_order, ax=axis1)
monthbar.set_xlabel('Month')
monthbar.set_ylabel('Count of crimes')
monthbar.set_title('Avg crimes per month')
_ = plt.setp(monthbar.get_xticklabels(), rotation=90)
#=================================================
# boxplot, check for outliers
axis2=  plt.subplot2grid((2,2),(0,1))
monthbox = sns.boxplot(x='Month', y='Category', data=Month_count,palette="Blues",
                   order=month_order, ax=axis2)
monthbox.set_xlabel('Month')
monthbox.set_ylabel('Count of crimes')
_ = plt.setp(monthbox.get_xticklabels(), rotation=90)
#=================================================
axis3=  plt.subplot2grid((2,2),(1,0))
monthbox = sns.barplot(x='Year', y='Category', data=Month_count,palette="Reds_r",
                    ax=axis3)
monthbox.set_xlabel('Year')
monthbox.set_ylabel('Count of crimes')
monthbox.set_title('Avg per year')
_ = plt.setp(monthbox.get_xticklabels(), rotation=90)
#=================================================
axis4=  plt.subplot2grid((2,2),(1,1))
monthbox = sns.boxplot(x='Year', y='Category', data=Month_count,palette="Reds",
                    ax=axis4)
monthbox.set_xlabel('Year')
monthbox.set_ylabel('Count of crimes')
_ = plt.setp(monthbox.get_xticklabels(), rotation=90)
#=================================================
plt.tight_layout()
```
